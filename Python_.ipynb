## Graphing/Plotting

```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Line plot')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot with a red line
plt.plot(x, y, color='r')

# Add labels and title
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Line plot with color')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Create plot with different colors for each line
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='b')

# Add labels and title
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Line plot with multiple colors')

# Show plot
plt.show()
```
